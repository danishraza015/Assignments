Q5 Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(0)

# Generate synthetic data with correlated features
n_samples = 100
feature_1 = np.random.normal(0, 1, n_samples)
feature_2 = feature_1 * 0.8 + np.random.normal(0, 0.2, n_samples)
feature_3 = feature_1 * -0.6 + np.random.normal(0, 0.3, n_samples)
feature_4 = feature_2 * 0.5 + feature_3 * -0.2 + np.random.normal(0, 1, n_samples)

# Create DataFrame
data = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3,
    'Feature 4': feature_4
})

# Calculate correlation matrix
correlation_matrix = data.corr()

# Plot heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix Heatmap of Synthetic Dataset")
plt.show()
```

Q2 Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
import numpy as np
import pandas as pd

# Prepare the output to display within the notebook
output_notebook()

# Generate random data
np.random.seed(42)
num_points = 50
x = np.random.rand(num_points) * 100
y = np.random.rand(num_points) * 100
sizes = np.random.randint(10, 30, size=num_points)  # Random marker sizes
colors = np.random.choice(["blue", "green", "red", "purple", "orange"], size=num_points)  # Random colors

# Create a data source
data = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a scatter plot
p = figure(title="Scatter Plot with Random Sizes and Colors", x_axis_label='X', y_axis_label='Y')
p.scatter(x='x', y='y', size='sizes', color='colors', alpha=0.6, source=data)

# Display the plot
show(p)
```

Q3 Generate a Bokeh bar chart representing the counts of different fruits using the following dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource
from bokeh.io import push_notebook

# Prepare the data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create the figure
p = figure(x_range=fruits, height=400, title="Fruit Counts",
           toolbar_location=None, tools="")

# Add bars to the figure
p.vbar(x='fruits', top='counts', width=0.5, source=source, color="skyblue")

# Configure the chart
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.axis_label = "Fruit"
p.yaxis.axis_label = "Count"

# Display the chart in notebook
output_notebook()
show(p, notebook_handle=True)
```

Q5 Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColorBar, LinearColorMapper, ColumnDataSource
from bokeh.transform import transform
from bokeh.io import output_notebook

# Generate random data for the heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Prepare data for Bokeh
data_flat = data_heatmap.flatten()
x_flat = np.repeat(x, len(y))
y_flat = np.tile(y, len(x))

# Create a color mapper
mapper = LinearColorMapper(palette="Viridis256", low=data_flat.min(), high=data_flat.max())

# Create a Bokeh figure
p = figure(title="Heatmap Example", x_axis_label='X', y_axis_label='Y',
           tools="hover", tooltips="@x, @y: @value", width=600, height=600)

# Create a ColumnDataSource with the data
source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, value=data_flat))

# Add squares to the heatmap, referencing columns in the source
# The 'x', 'y', and 'value' arguments are now strings that correspond
# to the keys in the ColumnDataSource.
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
        line_color=None, fill_color=transform('value', mapper))

# Add color bar
color_bar = ColorBar(color_mapper=mapper, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the output in a notebook
output_notebook()
show(p)
```
