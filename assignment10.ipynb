  (vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).

  

  (vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of
outliers

````python
import random
from collections import Counter
import math
from typing import List, Tuple

# Generate a list of 100 integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# Function to calculate mean
def calculate_mean(data: List[int]) -> float:
    return sum(data) / len(data)

# Function to calculate median
def calculate_median(data: List[int]) -> float:
    sorted_data = sorted(data)
    n = len(sorted_data)
    mid = n // 2
    if n % 2 == 0:
        return (sorted_data[mid - 1] + sorted_data[mid]) / 2
    else:
        return sorted_data[mid]

# Function to calculate mode
def calculate_mode(data: List[int]) -> List[int]:
    frequency = Counter(data)
    max_freq = max(frequency.values())
    return [key for key, val in frequency.items() if val == max_freq]

# Function to calculate weighted mean
def calculate_weighted_mean(values: List[int], weights: List[int]) -> float:
    weighted_sum = sum(v * w for v, w in zip(values, weights))
    total_weights = sum(weights)
    return weighted_sum / total_weights

# Function to calculate geometric mean
def calculate_geometric_mean(data: List[int]) -> float:
    product = math.prod(data)
    n = len(data)
    return product ** (1 / n)

# Function to calculate harmonic mean
def calculate_harmonic_mean(data: List[int]) -> float:
    n = len(data)
    return n / sum(1 / x for x in data)

# Function to calculate midrange
def calculate_midrange(data: List[int]) -> float:
    return (min(data) + max(data)) / 2

# Function to calculate trimmed mean
def calculate_trimmed_mean(data: List[int], percentage: float) -> float:
    sorted_data = sorted(data)
    n = len(sorted_data)
    trim_count = int(n * percentage / 100)
    trimmed_data = sorted_data[trim_count : n - trim_count]
    return sum(trimmed_data) / len(trimmed_data)

# Compute all requested statistics
mean = calculate_mean(int_list)
median = calculate_median(int_list)
mode = calculate_mode(int_list)
weights = [random.randint(1, 5) for _ in range(100)]  # Generate random weights for demonstration
weighted_mean = calculate_weighted_mean(int_list, weights)
geometric_mean = calculate_geometric_mean(int_list)
harmonic_mean = calculate_harmonic_mean(int_list)
midrange = calculate_midrange(int_list)
trimmed_mean = calculate_trimmed_mean(int_list, 10)  # Trim 10% of outliers

(
    mean,
    median,
    mode,
    weighted_mean,
    geometric_mean,
    harmonic_mean,
    midrange,
    trimmed_mean,
)
````

Q3 Write a Python class representing a discrete random variable with methods to calculate its expected value and variance

````python
class DiscreteRandomVariable:
    def __init__(self, values: list, probabilities: list):
        """
        Initializes a discrete random variable.

        :param values: A list of possible values of the random variable.
        :param probabilities: A list of probabilities corresponding to the values.
        """
        if len(values) != len(probabilities):
            raise ValueError("Values and probabilities must have the same length.")
        if not math.isclose(sum(probabilities), 1.0):
            raise ValueError("Probabilities must sum to 1.")
        self.values = values
        self.probabilities = probabilities

    def expected_value(self) -> float:
        """
        Calculates the expected value (mean) of the random variable.

        :return: The expected value.
        """
        return sum(v * p for v, p in zip(self.values, self.probabilities))

    def variance(self) -> float:
        """
        Calculates the variance of the random variable.

        :return: The variance.
        """
        mean = self.expected_value()
        return sum(p * (v - mean) ** 2 for v, p in zip(self.values, self.probabilities))


# Example usage
values = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]

random_variable = DiscreteRandomVariable(values, probabilities)
expected_val = random_variable.expected_value()
variance_val = random_variable.variance()

expected_val, variance_val
````

Q4  Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes.

````python
import random

class FairDieSimulation:
    def __init__(self, sides: int = 6):
        """
        Initializes the simulation for a fair die.

        :param sides: The number of sides on the die (default is 6).
        """
        self.sides = sides
        self.values = list(range(1, sides + 1))
        self.probabilities = [1 / sides] * sides

    def roll_die(self, rolls: int = 1000) -> list:
        """
        Simulates rolling the die a given number of times.

        :param rolls: The number of rolls to simulate (default is 1000).
        :return: A list of outcomes from the rolls.
        """
        return [random.choice(self.values) for _ in range(rolls)]

    def expected_value(self) -> float:
        """
        Calculates the theoretical expected value of the die.

        :return: The expected value.
        """
        return sum(v * p for v, p in zip(self.values, self.probabilities))

    def variance(self) -> float:
        """
        Calculates the theoretical variance of the die.

        :return: The variance.
        """
        mean = self.expected_value()
        return sum(p * (v - mean) ** 2 for v, p in zip(self.values, self.probabilities))


# Simulate the rolling of a fair six-sided die
die_simulation = FairDieSimulation()
outcomes = die_simulation.roll_die(rolls=1000)

# Calculate theoretical expected value and variance
expected_val_theoretical = die_simulation.expected_value()
variance_theoretical = die_simulation.variance()

expected_val_theoretical, variance_theoretical
````

Q5 Create a Python function to generate random samples from a given probability distribution (e.g.,binomial, Poisson) and calculate their mean and variance

````python
import numpy as np

def generate_samples_and_calculate_stats(distribution: str, params: dict, size: int = 1000) -> dict:
    """
    Generates random samples from a given probability distribution and calculates their mean and variance.

    :param distribution: The type of distribution ('binomial', 'poisson', etc.).
    :param params: Parameters for the distribution (e.g., 'n' and 'p' for binomial, 'lam' for Poisson).
    :param size: The number of random samples to generate (default is 1000).
    :return: A dictionary containing the samples, mean, and variance.
    """
    if distribution == "binomial":
        samples = np.random.binomial(n=params["n"], p=params["p"], size=size)
    elif distribution == "poisson":
        samples = np.random.poisson(lam=params["lam"], size=size)
    else:
        raise ValueError(f"Unsupported distribution: {distribution}")

    mean = np.mean(samples)
    variance = np.var(samples, ddof=1)  # Sample variance

    return {
        "samples": samples,
        "mean": mean,
        "variance": variance,
    }

# Example usage for binomial distribution
binomial_params = {"n": 10, "p": 0.5}
binomial_stats = generate_samples_and_calculate_stats("binomial", binomial_params, size=1000)

# Example usage for Poisson distribution
poisson_params = {"lam": 3}
poisson_stats = generate_samples_and_calculate_stats("poisson", poisson_params, size=1000)

binomial_stats["mean"], binomial_stats["variance"], poisson_stats["mean"], poisson_stats["variance"]
````

Q7 Use seaborn library to load tips dataset. Find the following from the dataset for the columns total_bill
and tip`:

  

  (i) Write a Python function that calculates their skewness.

  (iii) Write a function that calculates the covariance between two columns.


  (iv) Implement a Python program that calculates the Pearson correlation coefficient between two columns.


  (v) Write a script to visualize the correlation between two specific columns in a Pandas DataFrame using
scatter plots.

````python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Function to calculate skewness
def calculate_skewness(column: pd.Series) -> float:
    """
    Calculates the skewness of a column.

    :param column: A pandas Series containing numeric values.
    :return: The skewness of the column.
    """
    return column.skew()

# Function to determine skewness type
def skewness_type(column: pd.Series) -> str:
    """
    Determines the type of skewness of a column.

    :param column: A pandas Series containing numeric values.
    :return: A string indicating the type of skewness.
    """
    skewness = calculate_skewness(column)
    if skewness > 0:
        return "Positive Skewness"
    elif skewness < 0:
        return "Negative Skewness"
    else:
        return "Approximately Symmetric"

# Function to calculate covariance between two columns
def calculate_covariance(column1: pd.Series, column2: pd.Series) -> float:
    """
    Calculates the covariance between two columns.

    :param column1: A pandas Series containing numeric values.
    :param column2: A pandas Series containing numeric values.
    :return: The covariance between the two columns.
    """
    return np.cov(column1, column2, ddof=1)[0, 1]

# Function to calculate Pearson correlation coefficient
def calculate_pearson_correlation(column1: pd.Series, column2: pd.Series) -> float:
    """
    Calculates the Pearson correlation coefficient between two columns.

    :param column1: A pandas Series containing numeric values.
    :param column2: A pandas Series containing numeric values.
    :return: The Pearson correlation coefficient.
    """
    return column1.corr(column2)

# Function to visualize correlation using scatter plot
def visualize_correlation(data: pd.DataFrame, column1: str, column2: str):
    """
    Visualizes the correlation between two columns using a scatter plot.

    :param data: A pandas DataFrame.
    :param column1: The name of the first column.
    :param column2: The name of the second column.
    """
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=data, x=column1, y=column2, alpha=0.7, color="blue")
    plt.title(f"Scatter Plot: {column1} vs {column2}")
    plt.xlabel(column1)
    plt.ylabel(column2)
    plt.grid(True)
    plt.show()

# Calculate skewness and covariance/correlation for the columns
total_bill_skewness = calculate_skewness(tips["total_bill"])
tip_skewness = calculate_skewness(tips["tip"])
total_bill_skew_type = skewness_type(tips["total_bill"])
tip_skew_type = skewness_type(tips["tip"])
covariance = calculate_covariance(tips["total_bill"], tips["tip"])
pearson_correlation = calculate_pearson_correlation(tips["total_bill"], tips["tip"])

# Visualize the correlation
visualize_correlation(tips, "total_bill", "tip")

(
    total_bill_skewness,
    tip_skewness,
    total_bill_skew_type,
    tip_skew_type,
    covariance,
    pearson_correlation,
)
````

Q8 Write a Python function to calculate the probability density function (PDF) of a continuous random variable for a given normal distribution.

````python
import math

def normal_pdf(x: float, mean: float, std_dev: float) -> float:
    """
    Calculates the probability density function (PDF) of a normal distribution.

    :param x: The value for which to calculate the PDF.
    :param mean: The mean of the normal distribution.
    :param std_dev: The standard deviation of the normal distribution.
    :return: The PDF value for the given x.
    """
    factor = 1 / (std_dev * math.sqrt(2 * math.pi))
    exponent = -((x - mean) ** 2) / (2 * (std_dev ** 2))
    return factor * math.exp(exponent)

# Example usage
x_value = 5.0
mean_value = 4.0
std_dev_value = 1.5

pdf_value = normal_pdf(x_value, mean_value, std_dev_value)
pdf_value
````

Q9 Create a program to calculate the cumulative distribution function (CDF) of exponential distribution

````python
import math

def exponential_cdf(x: float, lambda_param: float) -> float:
    """
    Calculates the cumulative distribution function (CDF) of an exponential distribution.

    :param x: The value for which to calculate the CDF (x >= 0).
    :param lambda_param: The rate parameter (lambda > 0).
    :return: The CDF value for the given x.
    """
    if x < 0:
        return 0.0  # CDF is 0 for negative x in exponential distribution
    return 1 - math.exp(-lambda_param * x)

# Example usage
x_value = 2.0
lambda_value = 1.5

cdf_value = exponential_cdf(x_value, lambda_value)
cdf_value
````

Q10 Write a Python function to calculate the probability mass function (PMF) of Poisson distribution

````python
import math

def poisson_pmf(k: int, lambda_param: float) -> float:
    """
    Calculates the probability mass function (PMF) of a Poisson distribution.

    :param k: The number of occurrences (k >= 0).
    :param lambda_param: The mean number of occurrences (lambda > 0).
    :return: The PMF value for the given k.
    """
    if k < 0:
        raise ValueError("k must be a non-negative integer.")
    return (lambda_param ** k * math.exp(-lambda_param)) / math.factorial(k)

# Example usage
k_value = 3
lambda_value = 2.5

pmf_value = poisson_pmf(k_value, lambda_value)
pmf_value
````

Q12 A tutoring service claims that its program improves students' exam scores. A sample of students who
participated in the program was taken, and their scores before and after the program were recorded.


Use the below code to generate samples of respective arrays of marks:

```python

before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])

after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

```

Use z-test to find if the claims made by tutor are true or false.

Ans To perform a z-test to compare the scores before and after the tutoring program, you would need to treat this as a paired sample z-test or a one-sample z-test for differences between two related groups. Here's the basic idea of how you can proceed:

Null Hypothesis (H0): The program does not improve the students' exam scores. Therefore, the difference between "after" and "before" scores has a mean of zero.
Alternative Hypothesis (H1): The program improves the students' exam scores, meaning the mean difference between "after" and "before" scores is greater than zero.
Step-by-Step Approach
Calculate the difference between the "after" and "before" scores for each student.
Compute the mean and standard deviation of these differences.
Perform the z-test based on the sample mean, sample standard deviation, and the sample size.
Here’s how you can implement this in Python:

Python Code to Perform Z-Test

````python
import numpy as np
from scipy.stats import norm

# Generate data
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Step 1: Calculate the differences
differences = after_program - before_program

# Step 2: Calculate the mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)  # sample standard deviation
n = len(differences)  # sample size

# Step 3: Perform the z-test
# Standard error of the mean difference
std_error = std_diff / np.sqrt(n)

# Z-statistic calculation
z_stat = mean_diff / std_error

# Step 4: Calculate the p-value (one-tailed test, since we're testing for improvement)
p_value = 1 - norm.cdf(z_stat)  # one-tailed test

# Results
z_stat, p_value
````

Explanation of Code:
Differences: The difference between the "after" and "before" scores for each student is calculated.
Mean & Standard Deviation: These are computed for the differences.
Z-Statistic: The z-statistic is computed by dividing the mean difference by the standard error of the differences.
P-Value: The p-value is obtained using the cumulative distribution function (CDF) for a one-tailed z-test.
Hypothesis Decision:
If p-value < 0.05, reject the null hypothesis (i.e., the tutoring program does improve scores).
If p-value >= 0.05, fail to reject the null hypothesis (i.e., there isn't enough evidence to conclude that the program improves scores).

Ans To test whether the new drug is effective in reducing blood pressure, you can use a paired sample z-test. Here’s how you can approach the problem:

Hypothesis
Null Hypothesis (H0): The drug does not reduce blood pressure, meaning the mean difference between the "before" and "after" measurements is zero.
Alternative Hypothesis (H1): The drug does reduce blood pressure, meaning the mean difference between "before" and "after" measurements is less than zero.
Steps:

Calculate the mean and standard deviation of these differences.
Perform the z-test based on the sample mean, sample standard deviation, and sample size.
Test the hypothesis by calculating the z-statistic and comparing it with the significance level (typically 0.05).
Python Code to Implement the Z-Test

````python
import numpy as np
from scipy.stats import norm

# Generate data
before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])
after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])

# Step 1: Calculate the differences
differences = before_drug - after_drug  # before - after because we are testing for a reduction

# Step 2: Calculate the mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)  # sample standard deviation
n = len(differences)  # sample size

# Step 3: Perform the z-test
# Standard error of the mean difference
std_error = std_diff / np.sqrt(n)

# Z-statistic calculation
z_stat = mean_diff / std_error

# Step 4: Calculate the p-value (one-tailed test, since we're testing for a reduction)
p_value = norm.cdf(z_stat)  # one-tailed test

# Results
z_stat, p_value
````

P-Value: We calculate the one-tailed p-value using the cumulative distribution function (CDF) of the z-statistic.
Hypothesis Decision:
If p-value < 0.05, we reject the null hypothesis (i.e., the drug is effective in reducing blood pressure).
If p-value >= 0.05, we fail to reject the null hypothesis (i.e., there isn't enough evidence to support that the drug reduces blood pressure).

Python Code to Perform the Independent t-Test:

````python
import numpy as np
from scipy import stats

def quality_comparison(version1_scores, version2_scores):
    # Sample sizes
    n_1 = len(version1_scores)
    n_2 = len(version2_scores)

    # Sample means
    mean_1 = np.mean(version1_scores)
    mean_2 = np.mean(version2_scores)

    # Sample standard deviations
    std_1 = np.std(version1_scores, ddof=1)
    std_2 = np.std(version2_scores, ddof=1)

    # Calculate the t-statistic
    t_stat = (mean_1 - mean_2) / np.sqrt((std_1**2 / n_1) + (std_2**2 / n_2))

    # Calculate degrees of freedom using Welch-Satterthwaite equation
    df = ((std_1**2 / n_1 + std_2**2 / n_2)**2) / \
         ((std_1**2 / n_1)**2 / (n_1 - 1) + (std_2**2 / n_2)**2 / (n_2 - 1))

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))  # Two-tailed test

    return t_stat, df, p_value

# Data for version 1 and version 2
version1_scores = [85, 88, 82, 89, 87, 84, 90, 88, 85, 86, 91, 83, 87, 84, 89, 86, 84, 88, 85, 86, 89, 90, 87, 88, 85]
version2_scores = [80, 78, 83, 81, 79, 82, 76, 80, 78, 81, 77, 82, 80, 79, 82, 79, 80, 81, 79, 82, 79, 78, 80, 81, 82]

# Perform quality comparison
t_stat, df, p_value = quality_comparison(version1_scores, version2_scores)

t_stat, df, p_value
````

Q21 A political analyst wants to determine if there is a significant association between age groups and voter
preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify
them into different age groups and candidate preferences. Perform a Chi-Square test to determine if
there is a significant association between age groups and voter preferences.


Use the below code to generate data:

```python

np.random.seed(0)

age_groups = np.random.choice([ 18 30 , 31 50 , 51+', 51+'], size=30)

voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

Ans To determine if there is a significant association between age groups and voter preferences, we can use the Chi-Square test of independence. This test is used to assess whether two categorical variables (in this case, age groups and voter preferences) are independent or whether there is a significant association between them.

Steps for the Chi-Square Test:
Create a contingency table that shows the frequency of occurrences for each combination of age group and voter preference.
Apply the Chi-Square test to the contingency table to determine if the observed frequencies are significantly different from the expected frequencies under the assumption of independence.
Hypotheses:
Null Hypothesis (H0): There is no significant association between age groups and voter preferences (i.e., they are independent).
Alternative Hypothesis (H1): There is a significant association between age groups and voter preferences (i.e., they are dependent).
Interpret the p-value:
If p-value < 0.05, reject the null hypothesis and conclude that there is a significant association.
If p-value >= 0.05, fail to reject the null hypothesis, meaning there is no significant association.
Code Implementation:

````python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generating random data for age groups and voter preferences
np.random.seed(0)

# Age groups and voter preferences
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Create a DataFrame to view the data
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

# Create a contingency table
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Perform Chi-Square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Output the results
chi2_stat, p_value, dof, expected
````

Explanation of the Code:
Data Generation: We use np.random.choice to generate random data for age groups and voter preferences. The size=500 argument indicates that there are 500 data points.

Contingency Table: pd.crosstab is used to create a contingency table from the two categorical variables: age groups and voter preferences. The table shows how the voters are distributed across different combinations of age groups and preferences.

Chi-Square Test: chi2_contingency from scipy.stats is used to perform the Chi-Square test. It returns:

chi2_stat: The Chi-Square test statistic,
p_value: The p-value for the test,
dof: The degrees of freedom,
expected: The expected frequencies based on the null hypothesis.
Hypothesis Testing:
If p-value < 0.05, we reject the null hypothesis and conclude that there is a significant association between age groups and voter preferences.
If p-value >= 0.05, we fail to reject the null hypothesis, meaning there is no significant association.
Output:
The test will return the Chi-Square statistic, p-value, degrees of freedom, and the expected frequencies, allowing you to make an informed decision about the association between age groups and voter preferences.
